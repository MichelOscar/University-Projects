### Exercici 1

Si volem que la circumferència tingui radi 100, lògicament, les dues coordenades (x, y) han d'estar a 200 unitats de la seva coordenada corresponent.

```python
import tkinter  # Importem la llibreria.
root = tkinter.Tk()
finestra = tkinter.Canvas(root,  # Creem el llenç tal com estava definit a la pràctica.
borderwidth = 20,
background = 'white',
cursor = 'circle',
height = 240,
width = 320)
finestra.pack()
finestra.create_oval(50, 50,  # Creem la circumferència i li donem les coordenades segons el que hem definit abans.
    250, 250,
    fill = 'red',
    width = 8)  # Com gruixuda és un paràmetre prou relatiu, entenem 8 com a tal.
finestra.mainloop()  # Finim el programa de manera que es pugui tancar la finestra.
```

### Exercici 2

Per a fer això, haurem de situar el centre a (x, y), i ho farem definint-les com a origen i afegint o restant la meitat de mida (o sigui, el radi) a cada una d'elles. Els tres vèrtexs del nas els allunyarem del centre segons una fracció de mida, i seguim el mateix procés amb els ulls.

```python
def dibuixa_cara(finestra, x, y, mida):
        finestra.create_oval(x - mida/2, y - mida/2,  # Creem la circumferència que farà de cara pròpiament dita.
            x + mida/2, y + mida/2,
            width = 3)  # La fem més fina, per a que sigui una mica més realista.
        finestra.create_polygon(x,y - mida/8, x + mida/8,y, x - mida/8,y,  # Creem el nas i el situem al centre.
            outline='black',
            fill = 'black',  # El fem negre, com a l'exemple.
            width = 3)
        finestra.create_oval(x - 3*mida/8, y - 3*mida/8,  # Creem la circumferència que farà d'ull dret
            x - mida/8, y - mida/8,                       # (vist des de la pantalla).
            width = 3)  # La fem més fina, per a que sigui una mica més realista.
        finestra.create_oval(x + 3*mida/8, y - 3*mida/8,  # Creem la circumferència que farà d'ull esquerre
            x + mida/8, y - mida/8,                       # (vist des de la pantalla).
            width = 3)  # La fem més fina, per a que sigui una mica més realista.
        finestra.create_arc(x - mida/4, y + mida/8,  # Donem les coordenades el·líptiques de l'arc.
            x + mida/4, y + mida/3,
            fill = 'red',
            outline = 'red',
            width = 5,
            start = 180,  # Definim els angles d'inici i final.
            extent = 180,
            style = tkinter.ARC)
import tkinter  # Importem la llibreria.
root = tkinter.Tk()
window = tkinter.Canvas(root,  # Creem el llenç tal com estava definit a la pràctica.
    borderwidth = 20,
    background = 'white',
    cursor = 'circle',
    height = 240,
    width = 320)
window.pack()
dibuixa_cara(window, int(input("Quina és la coordenada x del centre? ")), 
             int(input("\nI la coordenada y? ")), int(input("\nI el diàmetre de la cara? ")))
window.mainloop()
```

### Exercici 3

Per a aquest exercici, importem la biblioteca random dins un for que repeteixi deu vegades el procés. A més, afegim dues variables més a la funció per al gruix de les línies i els colors. He suposat que el gruix corresponia per igual a totes les línies i que el color només havia d'afectar a la boca.

```python
def dibuixa_cara(finestra, x, y, mida, gruix, color):
        finestra.create_oval(x - mida/2, y - mida/2,  # Creem la circumferència que farà de cara pròpiament dita.
            x + mida/2, y + mida/2,
            width = gruix)  # La fem més fina, per a que sigui una mica més realista.
        finestra.create_polygon(x,y - mida/8, x + mida/8,y, x - mida/8,y,  # Creem el nas i el situem al centre.
            outline='black',
            fill = 'black',  # El fem negre, com a l'exemple.
            width = gruix)
        finestra.create_oval(x - 3*mida/8, y - 3*mida/8,  # Creem la circumferència que farà d'ull dret
            x - mida/8, y - mida/8,                       # (vist des de la pantalla).
            width = gruix)  # La fem més fina, per a que sigui una mica més realista.
        finestra.create_oval(x + 3*mida/8, y - 3*mida/8,  # Creem la circumferència que farà d'ull esquerre
            x + mida/8, y - mida/8,                       # (vist des de la pantalla).
            width = gruix)  # La fem més fina, per a que sigui una mica més realista.
        finestra.create_arc(x - mida/4, y + mida/8,  # Donem les coordenades el·líptiques de l'arc.
            x + mida/4, y + mida/3,
            fill = gba,
            outline = gba,
            width = gruix,
            start = 180,  # Definim els angles d'inici i final.
            extent = 180,
            style = tkinter.CHORD)
import tkinter  # Importem la llibreria.
root = tkinter.Tk()
window = tkinter.Canvas(root,  # Creem el llenç tal com estava definit a la pràctica.
    borderwidth = 20,
    background = 'white',
    cursor = 'circle',
    height = 240,
    width = 320)
window.pack()
for x in range (0, 10):
    import random
    vga = random.randint(1, 5)  # Assignem valors als colors.
    if vga == 1:
        gba = 'red'
    elif vga == 2:
        gba = 'blue'
    elif vga == 3:
        gba = 'green'
    elif vga == 4:
        gba = 'yellow'
    else:
        gba = 'black'
    dibuixa_cara(window, random.randint(0, 320), random.randint(0, 240), random.randint(5, 100), random.randint(1, 10), gba)
window.mainloop()
```

### Exercici 4

He pensat que era més clar posar els números a les marques de les hores enlloc de les marques suggerides. Per ajustar la seva posició he col·locat els quatre extrems (les 12, 3, 6 i 9) i he col·locat les hores intermitges fent servir funcions trigonomètriques, que, curiosament, he extret de les agulles que he fet posteriorment. També he hagut de passar els segons, minuts i hores a radiants i fer servir funcions trigonomètriques per determinar la longitud de l'agulla en cada una de les dues coordenades.

```python
from datetime import datetime
ara = datetime.now()
import tkinter  # Importem la llibreria.
root = tkinter.Tk()
finestra = tkinter.Canvas(root,  # Creem el llenç tal com estava definit a la pràctica.
    borderwidth = 20,
    background = 'white',
    cursor = 'circle',
    height = 240,
    width = 320)
finestra.pack()
finestra.create_oval(50, 50,  # Creem l'esfera del rellotge.
       250, 250,
       width = 3)
finestra.create_text(150, 60,  # Situem les marques de les 12 hores.
        text = "12")
finestra.create_text(150 + 90*math.sin((5/30)*math.pi), 150 - 90*math.cos((5/30)*math.pi),
        text = "1")
finestra.create_text(150 + 90*math.sin((10/30)*math.pi), 150 - 90*math.cos((10/30)*math.pi),
        text = "2")
finestra.create_text(240, 150,
        text = "3")
finestra.create_text(150 + 90*math.sin((20/30)*math.pi), 150 - 90*math.cos((20/30)*math.pi),
        text = "4")
finestra.create_text(150 + 90*math.sin((25/30)*math.pi), 150 - 90*math.cos((25/30)*math.pi),
        text = "5")
finestra.create_text(150, 240,
        text = "6")
finestra.create_text(150 + 90*math.sin((35/30)*math.pi), 150 - 90*math.cos((35/30)*math.pi),
        text = "7")
finestra.create_text(150 + 90*math.sin((40/30)*math.pi), 150 - 90*math.cos((40/30)*math.pi),
        text = "8")
finestra.create_text(60, 150,
        text = "9")
finestra.create_text(150 + 90*math.sin((50/30)*math.pi), 150 - 90*math.cos((50/30)*math.pi),
        text = "10")
finestra.create_text(150 + 90*math.sin((55/30)*math.pi), 150 - 90*math.cos((55/30)*math.pi),
        text = "11")
import math
finestra.create_line(150, 150,  # Situem l'inici del secunder a l'origen de coordenades del rellotge.
        150 + 90*math.sin(((ara.second)/30)*math.pi), 150 - 90*math.cos(((ara.second)/30)*math.pi),  # Situem el final depenent
        width = 1)                                                                                   # del temps.
finestra.create_line(150, 150,  # Situem l'inici de l'agulla de la minutera a l'origen de coordenades del rellotge.
        150 + 75*math.sin(((ara.minute)/30)*math.pi), 150 - 75*math.cos(((ara.minute)/30)*math.pi),  # Situem el final depenent
        width = 3)                                                                                   # del temps.
finestra.create_line(150, 150,  # Repetim els processos anteriors per l'horària.
        150 + 35*math.sin(((ara.hour)/6)*math.pi), 150 - 35*math.cos(((ara.hour)/6)*math.pi),
        width = 5)
finestra.mainloop()  # Finim el programa de manera que es pugui tancar la finestra.
```
