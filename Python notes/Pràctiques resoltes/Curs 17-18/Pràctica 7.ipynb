### Pas 1: *Definició d'una funció d'entrada de dades*

```python
def demanar_enter(text, min, max):  # Definim la funció
    n = int(input(text))  # Creem la variable per al nombre que demanem.
    while n > max or n < min:  # Si n no és a l'interval, la tornem a demanar molt educadament.
        n = int(input("Si li plau, cenyeixi's al rang establert i al conjunt de nombres enters. "))
    return n  # Retornem el valor d'n, ara sí, dins l'interval establert.

m = demanar_enter("Digui un número de l'1 al 5: ", 1, 5)  # Comprovem que la funció funcioni.
print (m)
```

### Pas 3: *Traçat de la graella*

```python
def ini_finestra():  # Definim la funció per crear la finestra que hem creat abans, per tal de poder provar la nova funció.
    root = tkinter.Tk()
    finestra = tkinter.Canvas(root,
        borderwidth = 20,
        background = 'yellow',
        cursor = 'circle',
        height = 600,
        width = 600)
    finestra.pack()
    return finestra

def ini_graella(window):  # Definim la nova funció, per crear la graella, amb l'argument de la finestra on dibuixarem.
    for n in range(0, 3):  # Creem un primer bucle, per a que es crein les tres columnes.
        for m in range(0, 3):  # Creem un segon bucle, per a que es crein les tres caselles de cada columna.
            window.create_polygon (75 + 150*n, 75 + 150*m,  # Ajustem totes les variables per tal que cada casella nova
                75 + 150*(n+1), 75 + 150*m,                 # medeixi 150x150 píxels i coincideixi amb el final de la
                75 + 150*(n+1), 75 + 150*(m+1),             # casella anterior, en marcar-ne tots quatre extrems.
                75 + 150*(n), 75 + 150*(m+1),
                outline = 'black',  # Fem que la vora sigui negra, per a que ressalti bé.
                fill = 'yellow',  # Ens assegurem que el farciment segueixi sent groc.
                width = 10)  # Fem gruixuda la vora, per a que no es trenqui.
            
import tkinter  # Importem la llibreria per provar la funció.
finestra = ini_finestra()  # Creem la finestra.
ini_graella(finestra)  # Iniciem la nova funció amb la finestra de la vella com a argument.
finestra.mainloop()  # Ens assegurem que la finestra es pugui tancar bé.
```

### Pas 4: Marquem la casella

```python
def ini_finestra():  # Definim les dues velles funcions per tal de provar la nova.
    root = tkinter.Tk()
    finestra = tkinter.Canvas(root,
        borderwidth = 20,
        background = 'yellow',
        cursor = 'circle',
        height = 600,
        width = 600)
    finestra.pack()
    return finestra

def ini_graella(window):
    for n in range(0, 3):
        for m in range(0, 3):
            window.create_polygon (75 + 150*n, 75 + 150*m,
                75 + 150*(n+1), 75 + 150*m,
                75 + 150*(n+1), 75 + 150*(m+1), 
                75 + 150*(n), 75 + 150*(m+1),
                outline = 'black',
                fill = 'yellow',
                width = 10)
            
def marcar_casella(y, x, jugador, window):  # Definim la funció tal com s'especifica al guió.
    if jugador == 1:
        window.create_oval (90 + 150*(x-1), 90 + 150*(y-1),  # Si es tracta del jugador 1, la seva fitxa és un cercle.
            60 + 150*x, 60 + 150*y,          # N'ajustem totes les coordenades a les caselles creades anteriorment.
            outline = 'blue',  # Fem que la vora del cercle sigui blava, per distingir-la bé.
            fill = 'yellow',  # Fem que el farciment segueixi sent groc.
            width = 20)  # Fem la vora gruixuda, per a que destaqui.
    elif jugador == 2:
        window.create_line (90 + 150*(x-1), 90 + 150*(y-1),  # Si es tracta del jugador 2, la seva fitxa és una creu.
            60 + 150*x, 60 + 150*y,          # La creem fent dues línies amb les coordenades ajustades.
            fill = 'red',  # La fem vermella, per a que s'oposi al cercle.
            width = 20)  # Fem la vora gruixuda, per a que ressalti.
        window.create_line (60 + 150*x, 90 + 150*(y-1),  # Creem la segona línia, inversa a la primera.
            90 + 150*(x-1), 60 + 150*y,
            fill = 'red',
            width = 20)
        
import tkinter  # Importem la llibreria per provar la funció.
finestra = ini_finestra()  # Creem la finestra.
ini_graella(finestra)  # Iniciem la funció graella amb la finestra com a argument.
marcar_casella (1, 2, 1, finestra)  # Provem la funció dos cops, un per a cada jugador.
marcar_casella (2, 1, 2, finestra)
finestra.mainloop()  # Ens assegurem que la finestra es pugui tancar bé.
```

### Pas 6: Control bàsic del joc
He decidit no fer global la variable jugador ja que no he vist la necessitat per a fer-ho. En el seu lloc, la variable jugador s'assigna en clicar el ratolí en funció de la variable torn.

```python
def ini_finestra():  # Definim les tres velles funcions per tal de provar les dues noves.
    root = tkinter.Tk()
    finestra = tkinter.Canvas(root,
        borderwidth = 20,
        background = 'yellow',
        cursor = 'circle',
        height = 600,
        width = 600)
    finestra.pack()
    return finestra

def ini_graella(window):
    for n in range(0, 3):
        for m in range(0, 3):
            window.create_polygon (75 + 150*n, 75 + 150*m,
                75 + 150*(n+1), 75 + 150*m,
                75 + 150*(n+1), 75 + 150*(m+1), 
                75 + 150*(n), 75 + 150*(m+1),
                outline = 'black',
                fill = 'yellow',
                width = 10)
            
def marcar_casella(y, x, jugador, window):
    if jugador == 1:
        window.create_oval (90 + 150*(x-1), 90 + 150*(y-1),
            60 + 150*x, 60 + 150*y,
            outline = 'blue',
            fill = 'yellow',
            width = 20)
    elif jugador == 2:
        window.create_line (90 + 150*(x-1), 90 + 150*(y-1),
            60 + 150*x, 60 + 150*y,
            fill = 'red',
            width = 20)
        window.create_line (60 + 150*x, 90 + 150*(y-1),
            90 + 150*(x-1), 60 + 150*y,
            fill = 'red',
            width = 20)

def fi_del_joc(window):  # Creem una funció per a la fi del joc, ens servirà més endavant.
    lletra = font.Font(family = 'Consolas',  # Creem la font per al missatge.
            size = 70,
            weight = 'bold')
    window.create_text (300, 300,  # Creem el text, al mig del canvas, per indicar la fi del joc.
            fill = 'chartreuse',
            text= "Fi del joc",
            font = lletra)

def control_de_joc (event):  # Creem el controlador del joc.
    global torn  # Globalitzem la variable torn.
    if torn%2 != 0:
        jugador = 1
    elif torn%2 == 0:
        jugador = 2
    x_i = event.x  # Assignem la posició del ratolí a la x a una variable.
    cx = 0
    while x_i > 90:  # Assignem un comptador de caselles i el fem funcionar.
        x_i = x_i - 150
        cx = cx + 1
    y_i = event.y
    cy = 0
    while y_i > 90:  # Fem el mateix procés per a la y.
        y_i = y_i - 150
        cy = cy + 1
    if cx < 1 or cx > 3:  # Evitem que sigui possible sortir de les caselles del tauler.
        torn = torn - 1
    elif cy < 1 or cy > 3:
        torn = torn - 1
    elif torn <= 9:
        marcar_casella(cy, cx, jugador, finestra)  # Cridem la funció per marcar la casella.
    if torn == 9:
        fi_del_joc(finestra)  # Acabem el joc al torn 9.
    torn = torn + 1  # Passem el torn.
    
import tkinter  # Importem la llibreria per provar la funció.
from tkinter import font  # Importem les fonts de tkinter.
finestra = ini_finestra()  # Creem la finestra.
ini_graella(finestra)  # Iniciem la funció graella amb la finestra com a argument.
torn = 1  # Iniciem la variable torn.
finestra.bind('<ButtonRelease-1>', control_de_joc)  # Assignem el click esquerre al control del joc.
finestra.mainloop()  # Ens assegurem que la finestra es pugui tancar bé.
```

### Pas 7: Millora! Control de caselles
Tant en aquest pas com en el 8, només comentem aquelles parts que són substancialment diferents al programa original, per no sobrecarregar el lector.

```python
def ini_finestra():
    root = tkinter.Tk()
    finestra = tkinter.Canvas(root,
        borderwidth = 20,
        background = 'yellow',
        cursor = 'circle',
        height = 600,
        width = 600)
    finestra.pack()
    return finestra

def ini_graella(window):
    for n in range(0, 3):
        for m in range(0, 3):
            window.create_polygon (75 + 150*n, 75 + 150*m,
                75 + 150*(n+1), 75 + 150*m,
                75 + 150*(n+1), 75 + 150*(m+1), 
                75 + 150*(n), 75 + 150*(m+1),
                outline = 'black',
                fill = 'yellow',
                width = 10)
            
def marcar_casella(y, x, jugador, window):
    if jugador == 1:
        window.create_oval (90 + 150*(x-1), 90 + 150*(y-1),
            60 + 150*x, 60 + 150*y,
            outline = 'blue',
            fill = 'yellow',
            width = 20)
    elif jugador == 2:
        window.create_line (90 + 150*(x-1), 90 + 150*(y-1),
            60 + 150*x, 60 + 150*y,
            fill = 'red',
            width = 20)
        window.create_line (60 + 150*x, 90 + 150*(y-1),
            90 + 150*(x-1), 60 + 150*y,
            fill = 'red',
            width = 20)

def imprimir(window, tuxt, y, m, color):  # Adaptem l'antiga funció fi del joc per a tot tipus de missatges.
    lletra = font.Font(family = 'Consolas',
            size = m,
            weight = 'bold')
    window.create_text (300, y,
            fill = color,
            text= tuxt,
            font = lletra)

def control_de_joc (event):
    global torn
    global tauler  # Fem global la llista de caselles ocupades.
    if torn%2 != 0:
        player = 1
    elif torn%2 == 0:
        player = 2
    x_i = event.x
    cx = 0
    while x_i > 90:
        x_i = x_i - 150
        cx = cx + 1
    y_i = event.y
    cy = 0
    while y_i > 90:
        y_i = y_i - 150
        cy = cy + 1
    if cx < 1 or cx > 3:
        torn = torn - 1
    elif cy < 1 or cy > 3:
        torn = torn - 1
    elif tauler[cx-1][cy-1] != 0:  # Impedim que el sistema permeti omplir dos cops la mateixa casella.
        torn = torn - 1
        imprimir(finestra, "No feu trampes!", 550, 20, 'red')
    elif torn <= 9:
        tauler[cx-1][cy-1] = player  # Fem que, cada cop que es cliqui a una casella, s'afegeixi a la llista.
        marcar_casella(cy, cx, player, finestra)
    if torn == 9:
        imprimir(finestra, "Fi del joc", 300, 70, 'chartreuse')
    torn = torn + 1
    
import tkinter
from tkinter import font
finestra = ini_finestra()
ini_graella(finestra)
torn = 1
tauler = [[0, 0, 0], [0, 0, 0], [0, 0, 0]]  # Creem una llista per a les caselles ocupades.
finestra.bind('<ButtonRelease-1>', control_de_joc)
finestra.mainloop()
```

```python
def ini_finestra():
    root = tkinter.Tk()
    finestra = tkinter.Canvas(root,
        borderwidth = 20,
        background = 'yellow',
        cursor = 'circle',
        height = 600,
        width = 600)
    finestra.pack()
    return finestra

def ini_graella(window):
    for n in range(0, 3):
        for m in range(0, 3):
            window.create_polygon (75 + 150*n, 75 + 150*m,
                75 + 150*(n+1), 75 + 150*m,
                75 + 150*(n+1), 75 + 150*(m+1), 
                75 + 150*(n), 75 + 150*(m+1),
                outline = 'black',
                fill = 'yellow',
                width = 10)
            
def marcar_casella(y, x, jugador, window):
    if jugador%2 != 0:
        window.create_oval (90 + 150*(x-1), 90 + 150*(y-1),
            60 + 150*x, 60 + 150*y,
            outline = 'blue',
            fill = 'yellow',
            width = 20)
    elif jugador%2 == 0:
        window.create_line (90 + 150*(x-1), 90 + 150*(y-1),
            60 + 150*x, 60 + 150*y,
            fill = 'red',
            width = 20)
        window.create_line (60 + 150*x, 90 + 150*(y-1),
            90 + 150*(x-1), 60 + 150*y,
            fill = 'red',
            width = 20)

def imprimir(window, tuxt, y, m, color):
    lletra = font.Font(family = 'Consolas',
            size = m,
            weight = 'bold')
    window.create_text (300, y,
            fill = color,
            text= tuxt,
            font = lletra)

def control_de_joc (event):
    global torn
    global tauler
    if torn%2 != 0:
        player = 1
    elif torn%2 == 0:
        player = 2
    x_i = event.x
    cx = 0
    while x_i > 90:
        x_i = x_i - 150
        cx = cx + 1
    y_i = event.y
    cy = 0
    while y_i > 90:
        y_i = y_i - 150
        cy = cy + 1
    if cx < 1 or cx > 3:
        torn = torn - 1
    elif cy < 1 or cy > 3:
        torn = torn - 1
    elif tauler[cx-1][cy-1] != 0 and torn <= 9:
        torn = torn - 1
        imprimir(finestra, "No feu trampes!", 550, 20, 'red')
    elif torn <= 9:
        tauler[cx-1][cy-1] = player
        marcar_casella(cy, cx, player, finestra)
        for y in range (0, 3):  # Activem, després de marcar la casella, un bucle que comprovi si aquell jugador ha guanyat.
            if tauler[0][y] == tauler[1][y] == tauler[2][y] != 0:
                imprimir(finestra, "Guanya el jugador " + str(player) + "!", 300, 38, 'chartreuse')
                torn = 10  # Trenquem el programa, fent que s'incompleixin totes les condicions per a que no es
                break      # pugui seguir jugant.
        for x in range (0, 3):  # Comprovem totes les possibilitats de victòria per a aquell jugador.
            if tauler[x][0] == tauler [x][1] == tauler [x][2] != 0:
                imprimir(finestra, "Guanya el jugador " + str(player) + "!", 300, 38, 'chartreuse')
                torn = 10
                break
        if tauler[0][0] == tauler[1][1] == tauler[2][2] != 0:  # Comprovem la diagonal principal.
            imprimir(finestra, "Guanya el jugador " + str(player) + "!", 300, 38, 'chartreuse')
            torn = 10
        elif tauler[2][0] == tauler[1][1] == tauler[0][2] != 0:  # Comprovem la diagonal secundària.
            imprimir(finestra, "Guanya el jugador " + str(player) + "!", 300, 38, 'chartreuse')
            torn = 10
    if torn == 9:
        imprimir(finestra, "Fi del joc", 300, 70, 'chartreuse')
    torn = torn + 1
    
import tkinter
from tkinter import font
finestra = ini_finestra()
ini_graella(finestra)
torn = 1
tauler = [[0, 0, 0], [0, 0, 0], [0, 0, 0]]
finestra.bind('<ButtonRelease-1>', control_de_joc)
finestra.mainloop()
```
