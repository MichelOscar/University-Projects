### Exercici 1.1:
Per a fer aquest exercici, fem servir la funció `random.random()` inclosa a la matriu, per generar una matriu `5x5` de nombres aleatoris, i les funcions `max()` i `min()` per localitzar els seus nombres major i menor.

```python
import numpy as np  # Importem la llibreria Numpy.
mat = np.random.random((5, 5))  # Creem la matriu de nombres aleatoris.
print(mat)  # Imprimim la matriu.
print("El major nombre de la matriu és: ", mat.max())  # Amb les funcions max i min, trobem el màxim i mínim de la matriu.
print("El menor nombre de la matriu és: ", mat.min())
```

### Exercici 1.3:
Per fer aquest exercici, definim una funció `valor(i, j)`, que ens retorni el valor demanat a l'enunciat, i establim que la matriu es generi a partir d'aquesta funció.

```python
def valor(i, j):  # Definim la funció.
    k = i**2 + 2*j  # Donem el valor a la k que retornarem.
    return k  # Retornem el valor de k.

import numpy as np  # Importem la llibreria.
mat = np.fromfunction(valor, (5, 5), dtype = int)  # Donem a la matriu els seus valors, segons la funció.
print(mat)  # Imprimim la matriu generada.
```

### Exercici 1.4:
Aquest exercici s'ha realitzat creant dos vectors amb valors aleatoris d'una fila i deu columnes, i fent pitàgores d'entre els dos. Per fer l'arrel simplement elevem a un mig. Calculem el màxim nombre del vector `R`, i així ja tenim el de major mòdul.

```python
import numpy as np  # Importem numpy.
vectorX = np.random.random((1, 10))  # Creem dos vectors de nombres aleatoris.
vectorY = np.random.random((1, 10))
R = (vectorX**2 + vectorY**2)**(1/2)  # Apliquem el teorema de pitàgores dels nombres d'ambdós per calcular-ne el mòdul.
print ("Les coordenades X són:\n", vectorX, "\nLes coordenades Y són:\n", vectorY, "\nEl mòdul d'aquests vectors és:\n", R)
print ("Per tant, el vector més llarg d'aquests medeix ", R.max(), ".")  # Imprimim els valors i trobem el màxim.
```

### Exercici 2: Gràfic bàsic amb *matplotlib*
Per fer aquest exercici, bàsicament, he dibuixat dues gràfiques superposades: una amb les línies discontínues en blau, i una altra amb una desena part dels punts marcats en blau.
A l'hora de fer aquest exercici m'he trobat bàsicament amb el problema d'ajustar els eixos de manera que el punt `(0,-1)` quedés a la intersecció d'ambdós. Això només m'era possible fent servir el comandament `plot.axis(x_0, x_f, y_0, y_f)`. D'altra banda, fent servir aquest comandament no sabia fer que els eixos estessin marcats en un intèrval d'1 punt i 0,5 punts respectivament, cosa que sí que sabia fer amb `plot.xticks()` i `plot.yticks()`. A l'hora de triar entre una i l'altra, he decidit que la darrera era la més similar al demanat.

```python
import numpy as np  # Importem numpy.
import matplotlib.pylab as plot  # Importem matplotlib.pylab.
vectorX = np.linspace(0, 2*np.pi, 100)  # Generem el vector per a les X, amb 100 valors entre 0 i 2pi.
vectorY = np.cos(2*vectorX - np.pi)/(vectorX + 1)  # Generem el vector Y per a la fórmula donada.
vectorX_i = np.linspace(0, 2*np.pi, 10)  # Repetim el procés per a 10 valors.
vectorY_i = np.cos(2*vectorX_i - np.pi)/(vectorX_i + 1)
plot.plot(vectorX, vectorY,  # Imprimim a la plot les línies discontínues.
          color = 'blue', linestyle = 'dashed', linewidth = 1)
plot.plot(vectorX_i, vectorY_i,  # Imprimim a la plot els punts, 1 de cada 10.
          linestyle = 'none',
          marker = 'o', markersize = 4, markerfacecolor = 'red')
plot.annotate("Sisè punt", xy = (vectorX_i[5], vectorY_i[5]), xytext = (2.2, -0.75), arrowprops = dict(facecolor = 'black'))
plot.title("Exercici 2")  # Creem la fletxa i el títol.
plot.xticks(np.arange(0, 2*np.pi, 1))  # Definim la llargada dels eixos i la separació d'etiquetes.
plot.yticks(np.arange(-1, 1.5, 0.5))
plot.xlabel("angle")  # Posem nom a l'eix x.
plot.ylabel("$\cos(2X - \pi)/(X + 1)$")  # Posem nom a l'ei y.
plot.show()  # Dibuixem la plot.
```

### Exercici 3: Figures de Lissajous
Per a fer les figures, hem seguit el mateix mètode usat en l'anterior exercici, però separant-les amb `subplot()` i separant cadascuna de les altres amb la funció `tight_layout()`.

```python
import numpy as np  # Importem numpy.
import matplotlib.pylab as plot  # Importem matplotlib.pylab.
t = np.linspace(0, 2*np.pi, 1000)  # Creem el vector t, amb 1000 punts.
vectorX = np.sin(4*t)  # Generem els primers vectors x i y, tal com s'indica.
vectorY = np.sin(3*t)
plot.xticks(np.arange(-1, 1.5, 0.5))  # Posem els eixos a lloc i amb la llargada necessària.
plot.yticks(np.arange(-1, 1.5, 0.5))
plot.subplot(1, 3, 1)  # Creem la primera subplot.
plot.plot(vectorX, vectorY,  # Hi posem la primera figura.
          color = 'red', linestyle = 'solid', linewidth = 1)
vectorX = np.sin(6*t)  # Repetim el procés per a la segona figura.
vectorY = np.sin(3*t)
plot.subplot(1, 3, 2)
plot.plot(vectorX, vectorY,
          color = 'red', linestyle = 'solid', linewidth = 1)
vectorX = np.sin(2*t)  # Repetim el procés per a la tercera figura.
vectorY = np.sin(5*t)
plot.subplot(1, 3, 3)
plot.plot(vectorX, vectorY,
          color = 'red', linestyle = 'solid', linewidth = 1)
plot.tight_layout()  # Separem adequadament les figures.
plot.show()  # Posem la plot creada en pantalla.
```

### Exercici 4: Arrels de funcions
Per a fer aquest exercici, hem generat tres funcions i hem trobat, per a elles, la primera arrel (comptant l'ordre des de menys infinit a més infinit). Les hem dibuixat i hem marcat l'arrel amb una fletxa.

El cas que val la pena comentar és el de la funció logaritme neperià (la `funció_2(x)`). Aquesta funció hem trobat que no es podia aplicar a un vector, o sigui que hem generat una llista normal fent servir un bucle `for` i l'hem convertit en una `array` per a que totes quedessin en el mateix format, que és el demanat.

```python
def funció_1(x):  # Definim la primera funció que estudiarem...
    return x**2
def funció_2(x):  # ...definim la segona...
    return math.log(x)
def funció_3(x):  # ... i la tercera.
    return (x+1)*(x+2)*(x+3)*(x+4)

import math  # Importem math.
import scipy.optimize as so  # Importem scipy.optimize.
import numpy as np  # Importem numpy.
import matplotlib.pylab as plot  # Importem matplotlib.
vectorX = np.linspace(-5.0, 5.0, 100)  # Generem el vectorX.
vectorX_2 = np.linspace(0.1, 5.0, 50)  # Generem un vectorX diferent, per a fer-lo compatible amb el logaritme.
vectorY_1 = funció_1(vectorX)  # Generem el primer vector i.
plot.plot(vectorX, vectorY_1)  # Generem la primera plot.
plot.annotate("Arrel", xy = (so.newton(funció_1, -5), 0), xytext = (0, 10), arrowprops = dict(facecolor = 'black'))
plot.title("$x^2$")  # Fem la fletxa i el títol de la primera plot.
plot.show()  # Ensenyem la primera plot.
print("Una arrel d'aquesta funció és a ", so.newton(funció_1, -5), " a l'eix de les x.")  # Imprimim la coordenada.
vectorY2 = []  # Creem el vector de la segona funció i fem el mateix procediment.
for y in range(0, 50):
    vectorY2.append(funció_2(vectorX_2[y]))
vectorY_2 = np.array(vectorY2)
plot.plot(vectorX_2, vectorY_2)
plot.annotate("Arrel", xy = (so.newton(funció_2, 0.1), 0), xytext = (0, 1), arrowprops = dict(facecolor = 'black'))
plot.title("$\lnx$")
plot.show()
print("Una arrel d'aquesta funció és a ", so.newton(funció_2, 0.2), "a l'eix de les x.")
vectorY_3 = (vectorX + 1)*(vectorX + 2)*(vectorX + 3)*(vectorX + 4)  # Repetim el procés per a la tercera plot.
plot.plot(vectorX, vectorY_3)
plot.annotate("Arrel", xy = (so.newton(funció_3, 0.1), 0), xytext = (0, 1000), arrowprops = dict(facecolor = 'black'))
plot.title("$(x+1)(x+2)(x+3)(x+4)$")
plot.show()
print("Una arrel d'aquesta última funció és a ", so.newton(funció_3, -5.), "a l'eix de les x.")
```
