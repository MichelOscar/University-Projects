###Exercic 1
$$ a=\frac{2x}{x+y} $$
$$ b=a^{x+y}-\left(\frac{z}{2x}\right)^2 $$
$$ c=\frac{2}{1+\frac{x}{y}} $$
$$ d=x^{(y+1)^{z+2}} $$
El següent programa demana els valors de x,y,z per teclat i dóna els resuñtats de les expressions anteriors.

```
# Demanem els valors de les variables x,y,z
x=float(input("x="))
y=float(input("y="))
z=float(input("z="))

# Realitzem les operacions per a calcular a,b,c,d evitant errors amb sentències if
# No he aconseguit evitar l'error overflow
if x==-y:
    a="No es pot fer"
else:
    a=(2**x)/(x+y)
if x==0 or int(int(a**(x+y)-(z/(2**x))**2).bit_length())>64:
    b="No es pot fer"
else:
    b=a**(x+y)-(z/(2**x))**2
if y==0 or (x/y)==-1:
    c="No es pot fer"
else:
    c=2/(1+x/y)
if int(int(x**((y+1)**(z+2))).bit_length())>64:
    d="No es pot fer"
else:
    d=x**((y+1)**(z+2))

# Imprimim els resultats
print("Per els valors de x={} y={} z={}\n\t a={}\n\t b={}\n\t c={}\n\t d={}".format(x,y,z,a,b,c,d))
```
