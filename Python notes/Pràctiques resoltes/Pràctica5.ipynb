###Exercici 1
Feu un programa que dibuixi un quadrat de 100 punts de costat tal que:
* El quadrat sigui vermell
* Tingui la vora gruixuda
* Estigui girat 45 graus
* Estigui centrat fora de l’origen

```python
# Finestra
wn=trt.Screen()
wn.setup(500,500)
wn.bgcolor("lightgreen")
wn.title("Exercici 1")

# Tortuga
alex=trt.Turtle()
alex.color("red")
alex.pensize(5)
alex.penup()
alex.goto(rd.randint(-150,150),rd.randint(-150,150))
alex.setheading(225)

# Dibuix
alex.pendown()
alex.forward(100)
alex.right(90)
alex.forward(100)
alex.right(90)
alex.forward(100)
alex.right(90)
alex.forward(100)


trt.exitonclick()
```

###Exercici 2
Feu un programa que dibuixi un polígon regular de qualsevol nombre de costats. El programa ha de
demanar per consola el nombre de costats i la mida del polígon i dibuixar-lo.

Opcional: Estructureu el programa de forma que el traçat del polígon estigui encapsulat en una funció: `dibuixa_poligon(tortuga, costats, mida)` on tortuga és la variable que heu creat a la tortuga.

```python
def dibuixa_poligon(tortuga,costats,mida):
    '''Dibuixa un polígon amb el nombre de costats i la mida d'aquests demanada'''
    tortuga.setheading(0)
    a=0
    while a<costats:
        tortuga.forward(mida)
        tortuga.left(180-(180*(costats-2))/costats) # Per a girar en l'angle que té el polígon regular 
                                                    # amb aquests costats
        a=a+1
    tortuga.hideturtle() # Amaga la tortuga
```

Funció alternativa amb la funció circle

```python
def dibuixa_poligon_c(tortuga,costats,mida):
    '''Dibuixa un polígon amb el nombre de costats i la mida d'aquests demanada'''
    tortuga.hideturtle()
    angle_poligon=m.radians((180*(costats-2))/costats) 
    tortuga.circle(mida/(2*m.cos(angle_poligon/2)),None,costats) # cos(angle/2)=mida/radi
```

```python
# Demanem les variables del polígon
costats=int(input("Nombre de costats del polígon: "))
mida=int(input("Tamany dels costats del polígon: "))

# Finestra
wn=trt.Screen()
wn.setup(mida*10,mida*10) # Adapta la pantalla a la mida del polígon
wn.bgcolor("lightgreen")
wn.title("Exercici 2")

# Tortuga
maria=trt.Turtle()
maria.color(rd.choice(["blue","red","yellow","purple","pink"])) # Color i traça aleatòries
maria.pensize(rd.randint(1,10))


dibuixa_poligon(maria,costats,mida)

trt.exitonclick()
```

Amb funció circle

```python
# Demanem les variables del polígon
costats=int(input("Nombre de costats del polígon: "))
mida=int(input("Tamany dels costats del polígon: "))

# Finestra
wn=trt.Screen()
wn.setup(mida*10,mida*10) # Adapta la pantalla a la mida del polígon
wn.bgcolor("lightgreen")
wn.title("Exercici 2.2")

# Tortuga
maria=trt.Turtle()
maria.color(rd.choice(["blue","red","yellow","purple","pink"])) # Color i traça aleatòries
maria.pensize(rd.randint(1,10))


dibuixa_poligon_c(maria,costats,mida)

trt.exitonclick()
```

###Exercici 3
Modifiqueu l’exercici anterior per fer un programa que dibuixi 10 polígons regulars de característiques
aleatòries: entre 3 i 10 costats, i mides entre 5 i 50 punts de costat. Aquesta modificació us resultarà
molt més fàcil si heu empaquetat la generació d’un polígon en una funció ``dibuixa_poligon()``

```python
# Finestra
wn=trt.Screen()
wn.setup(800,800)
wn.bgcolor("lightgreen")
wn.title("Exercici 3")

# Tortuga
pol=trt.Turtle()


# Bucle que executa l'acció 10 cops
n=0
while n<10:
    # Establim el color i el tamany aleatòris
    pol.color(rd.choice(["blue","red","yellow","purple","pink"]))
    pol.pensize(rd.randint(1,10))
    # Establim valors aleatòris de costats i mida d'aquests
    costats=rd.randint(3,10)
    mida=rd.randint(5,50)
    # Portem a la tortuga a qualsevol lloc de la finestra
    pol.penup()
    pol.goto(rd.randint(-300,300),rd.randint(-300,300))
    pol.pendown()
    # Executem la funció i actualitzem el bucle
    dibuixa_poligon(pol,costats,mida)
    n=n+1
    

trt.exitonclick()
```

###Exercici 4 (opcional)
Feu un programa amb turtle que representi l’hora actual en un rellotge analògic.

```python
# Variables temporals
ara=dt.datetime.now()
hara=ara.hour
minara=ara.minute
sara=ara.second

# Finestra
wn=trt.Screen()
wn.setup(800,800)
wn.title("Hora actual {:d}:{:d}:{:2f}".format(hara,minara,sara))

# Tortugues
t0=trt.Turtle() # rellotge
t0.penup()
t0.goto(0,-360)
t0.pendown()
t0.circle(360)
t0.hideturtle()

t1=trt.Turtle() # segons
t1.setheading(-sara*6+90) # L'ajust dels angles on apunten les manetes del rellotge és d'aquesta forma perquè
t1.forward(350)           # es mouen en sentit contrari que els angles i està girat 90 graus
t1.stamp()
t1.hideturtle()

t2=trt.Turtle() # minuts
t2.pensize(3)
t2.setheading(-minara*6+90)
t2.forward(310)
t2.hideturtle()

t3=trt.Turtle() # hores
t3.color("red")
t3.pensize(5)
t3.setheading(-hara*30+90)
t3.forward(270)
t3.hideturtle()


trt.exitonclick()
```
