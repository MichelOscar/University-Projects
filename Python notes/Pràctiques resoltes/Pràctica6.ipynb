```python
# Llibreries necessàries
import turtle as trt
import random as rd
```

####Rectangles

```python
def dibuixar_rectangle(tortuga,horitzontal,vertical):
    tortuga.setheading(0)
    tortuga.forward(horitzontal)
    tortuga.right(90)
    tortuga.forward(vertical)
    tortuga.right(90)
    tortuga.forward(horitzontal)
    tortuga.right(90)
    tortuga.forward(vertical)
```

###Ulls

```python
def dibuixar_ulls(tortuga,x,y,L):
    tortuga.penup()
    tortuga.goto(x+L/6,y+L/4-L/10)
    tortuga.pendown()
    tortuga.circle(L/10)
    tortuga.penup()
    tortuga.goto(x-L/6,y+L/4-L/10)
    tortuga.pendown()
    tortuga.circle(L/10)
```

###Nas

```python
def dibuixar_nas(tortuga,x,y,L):
    tortuga.penup()
    tortuga.goto(x,y)
    tortuga.pendown()
    tortuga.setheading(300)
    tortuga.forward(L/10)
    tortuga.right(120)
    tortuga.forward(L/10)
    tortuga.right(120)
    tortuga.forward(L/10)
```

###Boca

```python
def dibuixar_boca(tortuga,x,y,L):
    tortuga.penup()
    tortuga.goto(x-L/8,y-L/4)
    tortuga.pendown()
    tortuga.setheading(0)
    tortuga.forward(L/4)
```

###Cara

```python
def dibuixar_cara(tortuga,x,y,L):
    tortuga.penup()
    tortuga.goto(x,y-L/2)
    tortuga.pendown()
    tortuga.circle(L/2)
```

###Cap

```python
def dibuixar_cap(tortuga,x,y,L):
    dibuixar_cara(tortuga,x,y,L)
    dibuixar_ulls(tortuga,x,y,L)
    dibuixar_nas(tortuga,x,y,L)
    dibuixar_boca(tortuga,x,y,L)
```

###Moltes cares

```python
# Finestra
wn=trt.Screen()
wn.bgcolor("black")
wn.setup(1280,940)
wn.title("Moltes cares")

# Tortuga
alex=trt.Turtle()

for val in range (0,10): # El bucle s'executa 10 cops
    x=rd.randint(-565,565) # Posicions calculades per a què
    y=rd.randint(-395,395) # les cares no se surtin de la pantalla
    L=rd.randint(10,150)
    alex.speed(0)
    alex.color(rd.choice(["blue","red","yellow","pink","purple","orange","green","lightgreen","white"]))
    dibuixar_cap(alex,x,y,L)
    
alex.hideturtle()

trt.exitonclick()
```

###Cos

```python
def dibuixar_cos(tortuga,x,y,L):
    tortuga.penup()
    tortuga.goto(x-L/2,y-L/2)
    tortuga.pendown()
    dibuixar_rectangle(tortuga,L,2*L)
```

###Braços

```python
def dibuixar_bracos(tortuga,x,y,L):
    tortuga.penup()
    tortuga.goto(x-3*L/2,y-L/2)
    tortuga.pendown()
    dibuixar_rectangle(tortuga,L,L/8)
    tortuga.penup()
    tortuga.goto(x+L/2,y-L/2)
    tortuga.pendown()
    dibuixar_rectangle(tortuga,L,L/8)
```

###Cames

```python
def dibuixar_cames(tortuga,x,y,L):
    tortuga.penup()
    tortuga.goto(x-L/2,y-5*L/2)
    tortuga.pendown()
    dibuixar_rectangle(tortuga,L/8,L)
    tortuga.penup()
    tortuga.goto(x+3*L/8,y-5*L/2)
    tortuga.pendown()
    dibuixar_rectangle(tortuga,L/8,L)
```

###Ninot

```python
def dibuixar_ninot(tortuga,x,y,L):
    dibuixar_cap(tortuga,x,y,L)
    dibuixar_cos(tortuga,x,y,L)
    dibuixar_bracos(tortuga,x,y,L)
    dibuixar_cames(tortuga,x,y,L)
```

###Dibuix dels ninots

```python
# Finestra
wn=trt.Screen()
wn.bgcolor("black")
wn.setup(1280,940)
wn.title("Ninots")

# Tortuga
alex=trt.Turtle()

for val in range (0,5): # El bucle s'executa 5 cops
    x=rd.randint(-415,415) # Posicions calculades per a què
    y=rd.randint(-55,395)  # els ninots no se surtin de la pantalla
    L=rd.randint(10,150)
    alex.speed(0)
    alex.color(rd.choice(["blue","red","yellow","pink","purple","orange","green","lightgreen","white"]))
    dibuixar_ninot(alex,x,y,L)
    alex.setheading(0)
    
alex.hideturtle()

trt.exitonclick()
```
