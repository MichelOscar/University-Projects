### Exercici 1
**En aquest exercici farem un programa que compti les paraules d’un fitxer de text. Seguiu les especificacions
següents:**

* **Definiu una funcio que rebi una variable de tipus str i compti les paraules que conte. Una paraula es defineix com una serie de caràcters delimitats per: espais, signes de puntuacio, un principi de linia o un final de línia**

**Indicacio: per comptar les paraules d’una lınia compteu les vegades que es troba un caràcter normal seguit d’un caràcter de separacio (signe de puntuació, blanc o final de linia)**

Aquesta funció primer transforma a cadena l'argumant donat i defineix una llista amb els caràcters permesos.  
Després per cada element en la cadena si és un caràcter permès i el següent no ho és afegeix 1 a la variable words. Si no hi ha següent, afegeix 1 a words i retorna words.  
Finalment retorna words.

```python
def comptar_paraules(string):
    '''Compta el nombre de paraules que conté una cadena.'''
    
    # Conversió a str (per assegurar)
    string=str(string)
    
    # Caràcters que pot contenir una paraula
    minus=["a","b","c","ç","d","e","f","g","h","i","j","k","l","m","n","ñ","o","p","q","r","s","t","u","v","w","x","y","z"]
    mayus=["A","B","C","Ç","D","E","F","G","H","I","J","K","L","M","N","Ñ","O","P","Q","R","S","T","U","V","W","X","Y","Z"]
    voc_esp=["à","á","ä","è","é","ë","í","ì","ï","ò","ó","ö","ù","ú","ü","À","Á","Ä","È","É","Ë","Ì","Í","Ï","Ò","Ó","Ö","Ù","Ú","Ü"]
    num=[str(x) for x in range(10)]
    car_perm=minus+mayus+voc_esp+num
    
    # Iteració
    words=0
    for i in range(len(string)): # i serà l'índex
        if string[i] in car_perm: # Si ets un caràcter permès
            if i==len(string)-1: # Si ets l'últim, afegeix una paraula i retorna words
                words=words+1
                return words
            elif not string[i+1] in car_perm: # Si no ets l'últim i no ets un caràcter permès afegeix una paraula
                words=words+1
    
    return words

a="fgrfdt\ncvcxdf"
print(a, comptar_paraules(a),sep="\t")
```

**Feu un programa que:**
1. **Demani un nom de fitxer de text**
2. **Llegeixi el fitxer lınia a lınia**
3. **Per cada lınia usi la funcio anterior per comptar les paraules de la lınia**
4. **Acumuli el numero total de paraules i acabi el programa indicant el total**

Aquest programa demana un fitxer per pantalla. Després l'obre i compta les paraules amb la funció `comptar_paraules` definida en la cel·la anterior, les línies i el número de paraules.

```python
# Demanem el fitxer
fitxer=input("Nom del fitxer: ")

# L'obrim en mode llegir
with open(fitxer,'r') as CountingFile:
    paraules=0
    linies=0
    caracters=0
    for line in CountingFile.readlines(): # per cada ínia
        paraules=paraules+comptar_paraules(line) # compta les paraules
        linies=linies+1 # afegeix una línia
        caracters=caracters+len(line) # compta la quantitat de caràcters
        
print("Línies: {}\nParaules: {}\nCaràcters: {}".format(linies,paraules,caracters))
```

### Exercici 2
En aquest exercici farem un programa que escrigui un fitxer en format json i un altre que el llegeixi.

#### Primer programa
**Feu un primer programa que generi tres vectors:**
1. **Un vector amb valors de tipus int amb la seqüencia 0, 5, 15 , ..., 350**
2. **Un vector de tipus string on per cada angle s'indiqui el quadrant que li correspon:**

 *  **(0) "pz"**
 *  **(1,89) "pp"**
 *  **(90) "zp"**
 *  **(91,179) "np"**
 *  **(180) "nz"**
 *  **(181,269) "nn"**
 *  **(270) "zn"**
 *  **(271,359) "pn"**

3. **Un vector amb valors float on cada element contingui el cosinus dels angles de la primera seqüencia**
4. **Un vector amb valors float on cada element contingui el sinus dels angles de la primera seqüncia**
**El programa ha de guardar els vectors en un fitxer json**

Aquest programa guarda els vectors demanats, els imprimeix per a verificar posteriorment que retorna el mateix contingut, també imprimeix els tipus de dades.  
Finalment guarda les llistes en un arxiu json.

```python
import math as m

phi=[theta for theta in range(0,360,5)] # els angles del 0 al 355 amb itèrval de 5

quadrant=[]
for angle in phi: # segons el valor de l'angle afegeix a la llista una de les cadenes que indiquen el quadrant
    if angle==0:
        q="pz"
    elif angle in range(1,90):
        q="pp"
    elif angle==90:
        q="zp"
    elif angle in range(91,180):
        q="np"
    elif angle==180:
        q="nz"
    elif angle in range(181,270):
        q="nn"
    elif angle==270:
        q="zn"
    elif angle in range(271,360):
        q="pn"
    quadrant.append(q)

# Cosinus i sinus
cosinus=[m.cos(m.radians(theta)) for theta in phi]
sinus=[m.sin(m.radians(theta)) for theta in phi]

# Imprimim les llistes originals
print("{}\n{}\n{}\n{}".format(phi,quadrant,cosinus,sinus))

# I els tipus de dades originals
print("{}\n{}\n{}\n{}".format(type(phi[0]),type(quadrant[0]),type(cosinus[0]),type(sinus[0])))

import json

with open('angles.json','w') as jsonFile: # Obre l'arxiu en mode escriptura i guarda les llistes en un arxiu json
    json.dump({"angles":phi,"quadrant":quadrant,"cos":cosinus,"sin":sinus},jsonFile) # A cada llista se li assigana una
                                                                                     # clau per a poder recuperar-la
                                                                                     # posteriorment
```

#### Segon programa
**Feu un programa que llegeixi el fitxer generat pel primer i recuperi quatre vectors idèntics als anterior (el
tipus de dada i el contingut han de ser iguals)**

Aquest programa llegeix l'arxiu json creat anteriorment i retorna les llistes creades.

```python
import json

with open("angles.json",'r') as jsonFile: # Obre l'arxiu i retorna cada llista segons la clau donada anteriorment
    lists=json.load(jsonFile)
    phi=lists["angles"]
    quadrant=lists["quadrant"]
    cosinus=lists["cos"]
    sinus=lists["sin"]
    
# Imprimim les llistes
print("{}\n{}\n{}\n{}".format(phi,quadrant,cosinus,sinus))

# I els tipus de dades
print("{}\n{}\n{}\n{}".format(type(phi[0]),type(quadrant[0]),type(cosinus[0]),type(sinus[0])))
# Tots els elements passen pel mateix cast
```

Tant el contingut com el tipus de dada és el mateix.

### Exercici 3
**En aquest exercici farem un programa que escrigui un fitxer amb dades estructurades en format tabular i un
altre que el llegeixi.**  
**Mateixos programes que en l'exercici anterior però sense json.**

#### Programa 1
Aquest programa guarda els vectors demanats, els imprimeix per a verificar posteriorment que retorna el mateix contingut, també imprimeix els tipus de dades.  
Finalment guarda les llistes en un arxiu txt.

```python
import math as m

phi=[theta for theta in range(0,360,5)] # els angles del 0 al 355 amb itèrval de 5

quadrant=[]
for angle in phi: # segons el valor de l'angle afegeix a la llista una de les cadenes que indiquen el quadrant
    if angle==0:
        q="pz"
    elif angle in range(1,90):
        q="pp"
    elif angle==90:
        q="zp"
    elif angle in range(91,180):
        q="np"
    elif angle==180:
        q="nz"
    elif angle in range(181,270):
        q="nn"
    elif angle==270:
        q="zn"
    elif angle in range(271,360):
        q="pn"
    quadrant.append(q)

# Cosinus i sinus
cosinus=[m.cos(m.radians(theta)) for theta in phi]
sinus=[m.sin(m.radians(theta)) for theta in phi]

# Imprimim les llistes originals
print("{}\n{}\n{}\n{}".format(phi,quadrant,cosinus,sinus))

# I els tipus de dades originals
print("{}\n{}\n{}\n{}".format(type(phi[0]),type(quadrant[0]),type(cosinus[0]),type(sinus[0])))

with open("angles.txt",'w') as txtFile: # Obre l'arxiu en mode escriptura
    for i,j,k,l in zip(phi,quadrant,cosinus,sinus): # Per cada valor en el mateix índex en les llistes guarda'ls en un
        txtFile.write(str(i)+"\t"+                  # arxiu de text, en la mateixa línia, separats per un tabulador
                     j+"\t"+                        # i en el format especificat
                     "{:.4f}".format(k)+"\t"+
                     "{:.4f}".format(l)+"\n")
```

#### Programa 2
Aquest programa llegeix l'arxiu txt creat anteriorment i retorna les llistes creades.

```python
with open("angles.txt",'r') as txtFile: # Obre l'arxiu en mode lectura i fes una llista amb cadascuna de les línies
    lines=[i for i in txtFile.readlines()]
    
phi=[]
quadrant=[]
cosinus=[]
sinus=[]
for j in lines: # Per cada línia
    strip=j.strip() # treu l'últim caràcter
    split=strip.split("\t") # separa els elemants quant troba un tabulador
    phi.append(int(split[0])) # afegeix-los en la llista corresponent amb el tipus de dada que li correspon
    quadrant.append(split[1])
    cosinus.append(float(split[2]))
    sinus.append(float(split[3]))

# Imprimim les llistes
print("{}\n{}\n{}\n{}".format(phi,quadrant,cosinus,sinus))

# I els tipus de dades
print("{}\n{}\n{}\n{}".format(type(phi[0]),type(quadrant[0]),type(cosinus[0]),type(sinus[0])))
# Tots els elements passen pel mateix cast
```

El tipus de dades és el mateix.  
El contingut de les dues primeres llistes és el mateix però el de les que contenen floats els retorna amb només 4 decimals, ja que és com hem especificat que s'escrivissin.

#### Programa 1
Aquest programa guarda els vectors demanats, els imprimeix per a verificar posteriorment que retorna el mateix contingut, també imprimeix els tipus de dades.  
Finalment guarda les llistes en un arxiu csv, compatible amb formats d'arxiu nivell usuari i amb el format local.

```python
import math as m

phi=[theta for theta in range(0,360,5)] # els angles del 0 al 355 amb itèrval de 5

quadrant=[]
for angle in phi: # segons el valor de l'angle afegeix a la llista una de les cadenes que indiquen el quadrant
    if angle==0:
        q="pz"
    elif angle in range(1,90):
        q="pp"
    elif angle==90:
        q="zp"
    elif angle in range(91,180):
        q="np"
    elif angle==180:
        q="nz"
    elif angle in range(181,270):
        q="nn"
    elif angle==270:
        q="zn"
    elif angle in range(271,360):
        q="pn"
    quadrant.append(q)

# Cosinus i sinus
cosinus=[m.cos(m.radians(theta)) for theta in phi]
sinus=[m.sin(m.radians(theta)) for theta in phi]

# Imprimim les llistes originals
print("{}\n{}\n{}\n{}".format(phi,quadrant,cosinus,sinus))

# I els tipus de dades originals
print("{}\n{}\n{}\n{}".format(type(phi[0]),type(quadrant[0]),type(cosinus[0]),type(sinus[0])))

import locale as loc
loc.setlocale(loc.LC_ALL,"") # Permet usar el format local

with open("anglesCSV.csv",'w') as csvFile: # Obre l'arxiu en mode escriptura
    for i,j,k,l in zip(phi,quadrant,cosinus,sinus): # Per cada valor en el mateix índex en les llistes guarda'ls en un
        csvFile.write(str(i)+";"+                   # arxiu de csv, en la mateixa línia, separats per un ";"
                     j+";"+                         # i en el format local
                     "{:.4n}".format(k)+";"+
                     "{:.4n}".format(l)+"\n")
```

#### Programa 2
Aquest programa llegeix l'arxiu txt creat anteriorment i retorna les llistes creades.

```python
import csv
import locale as loc
loc.setlocale(loc.LC_ALL,"") # Permet usar el format local

lines=[]
with open("anglesCSV.csv",'r') as csvFile: # Obre l'arxiu en mode lectura
    reader=csv.reader(csvFile,delimiter=";",quotechar="\n")
    for line in reader: # Fes una llista amb totes les línies de l'arxiu
        lines.append(line)


phi=[loc.atoi(lines[i][0]) for i in range(len(lines))] # Recupera les llistes amb el format no local
quadrant=[lines[i][1] for i in range(len(lines))]
cosinus=[loc.atof(lines[i][2]) for i in range(len(lines))]
sinus=[loc.atof(lines[i][3]) for i in range(len(lines))]
    
# Imprimim les llistes
print("{}\n{}\n{}\n{}".format(phi,quadrant,cosinus,sinus))

# I els tipus de dades
print("{}\n{}\n{}\n{}".format(type(phi[0]),type(quadrant[0]),type(cosinus[0]),type(sinus[0])))
# Tots els elements passen pel mateix cast
```

El tipus de dades és el mateix.  
El contingut de les dues primeres llistes és el mateix però el de les que contenen floats els retorna amb només 4 decimals, ja que és com hem especificat que s'escrivissin.
