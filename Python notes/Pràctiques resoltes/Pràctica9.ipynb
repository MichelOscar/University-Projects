### Exercici 1
Matriu de 10x10 amb valors aleatoris.  
Trobar-ne màxim i mínim.

```python
import numpy as np

n=10
L=50
A=np.random.randint(-L,L+1,(n,n))
print(A)
print("Màxim=",A.max())
print("mínim=",A.min())
```

Matriu 5x5 normalitzada.

```python
import numpy as np

n=5
L=50
A=np.random.randint(-L,L+1,(n,n))
print("Matriu no normalitzada\n",A,sep="")
N=(A-A.min())/(A.max()-A.min())
print("Matriu normalitzada\n",N,sep="")
```

Dos vectors X,Y de 10 elements.  
Un vector amb les seves distàncies de punts amb coordenades x,y pertanyents a X,Y.  
La distància màxima.

```python
import numpy as np

L=50
X=np.random.randint(-L,L+1,(1,10))
print("X",X)
Y=np.random.randint(-L,L+1,(1,10))
print("Y",Y)
R=np.sqrt(X**2+Y**2)
print("R",R,"\n","Distància màxima",R.max())
```

### Exercici 2
Representació gràfica:
$$ f(x)=\frac{sin(2x-\pi)}{x+1} $$

```python
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt

def f(x):
    y=np.sin(2*x-np.pi)/(x+1)
    return y

x=np.linspace(0,2*np.pi,100)
y=f(x)

plt.xlim(0,2*np.pi)
plt.ylim(-1,1)
plt.title("Exercici 2")
plt.xlabel("angle")
plt.ylabel("$sin(2x-\pi)/(x+1)$")
P_6=np.pi
plt.annotate("Sisè punt",xy=(P_6,f(P_6)),xytext=(2,-0.5),
             arrowprops=dict(facecolor="black"))
plt.plot(x,y,color="blue",linestyle="dashed",marker="o",markerfacecolor="red",markevery=10)
```

### Exercici 3
Figures de Lissajus:
* Tenen com a coordenades:
$ x=Asin(\alpha t) $,
$ y=Bsin(\beta t) $  
$ t \in (0,2\pi) $

```python
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt

t=np.linspace(0,2*np.pi,1000)
def Lissajous(a,b):
    x=np.sin(a*t)
    y=np.sin(b*t)
    return x,y

alpha=[3,5,2]
beta=[4,4,7]
for a,b,i in zip(alpha,beta,range(3)):
    plt.subplot(1,3,i+1)
    x,y=Lissajous(a,b)
    plt.plot(x,y,'r-')
```

### Exercici 4
Arrels de funcions.  
Llibreria scipy

```python
import scipy.optimize as so
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt

def f(x):
    y=np.cos(x)
    return y
def g(x):
    y=1.5*x**2+2*x-5
    return y
def h(x):
    y=np.sin(x)
    return y
    
x=np.linspace(-0.5,2,1000)
for i in [f,g,h]:
    y=i(x)
    plt.plot(x,y,'r-')
    arrel=so.newton(i,1)
    plt.annotate("Arrel",xy=(arrel,i(arrel)),xytext=(0.5,-2),arrowprops=dict(facecolor="black"))
    
y=np.zeros((1000,1))
plt.plot(x,y,'g--')
```
