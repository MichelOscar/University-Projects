```python
%pylab inline
%config InlineBackend.figure_format = 'svg'
import numpy as np
import matplotlib.pylab as plt
x = np.linspace(-np.pi, np.pi, 10)
x1 = np.linspace(-np.pi, np.pi, 100)
y = np.sin(x)
y+=np.random.normal(0, 0.3, size=x.shape)
y1=np.sin(x1)
plt.plot(x,y, 'bh')
plt.plot(x1,y1, 'g')
plt.show()
```
